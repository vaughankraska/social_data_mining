```python
from text_mining.utils import preprocess_tweets
from sklearn.feature_extraction.text import CountVectorizer
from collections import Counter
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
import re
```

```python
def remove_urls(txt):
    return re.sub(r'https://t\.co/\w+', '', txt)
```

```python
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = re.sub(r'\d+', '', text)  # Remove numbers
    words = text.split()  # Tokenize
    words = [w for w in words if w not in ENGLISH_STOP_WORDS and len(w) > 2 and 'http' not in w]  # Remove stop words
    return words
```

```python
df["tweet_text_wo_url"] = df["tweet_text"].apply(remove_urls)
df['text'] = df['tweet_text_wo_url'].apply(preprocess_text)
df.head(10)
```

```python
ba_tweets = df[df["account_type"] == "Business actors"]["text"]
pi_tweets = df[df["account_type"] == "Private individuals"]["text"]

ba_word_counts = Counter([word for tweet in ba_tweets for word in tweet])
pi_word_counts = Counter([word for tweet in pi_tweets for word in tweet])

# print(ba_word_counts)
```

We are going to create a -1 to 1 scale where -1 represents Private individuals and 1 denotes Business actors 

```python
vocabulary = set(ba_word_counts.keys()).union(set(pi_word_counts.keys()))

word_scores = {}
total_ba_words = sum(ba_word_counts.values())
total_pi_words = sum(pi_word_counts.values())

for word in vocabulary:
    # ba_freq = ba_word_counts[word] if word in ba_word_counts and ba_word_counts[word] > 70 else 0
    # pi_freq = pi_word_counts[word] if word in pi_word_counts and pi_word_counts[word] > 70 else 0
    ba_freq = ba_word_counts.get(word, 0) / total_ba_words
    pi_freq = pi_word_counts.get(word, 0) / total_pi_words
    if ba_freq + pi_freq > 0:
        word_scores[word] = (ba_freq - pi_freq) / (ba_freq + pi_freq)
        
print(word_scores)
```

```python
n = 10

ba_top_words = sorted(word_scores.items(), key=lambda x: x[1], reverse=True)[:n]
pi_top_words = sorted(word_scores.items(), key=lambda x: x[1])[:n]

print("Top 10 words for Business actors:")
for word, score in ba_top_words:
    print(f"{word}: {score}")

print("\n---------------------------------\n")

print("\nTop 10 words for Private individuals:")
for word, score in pi_top_words:
    print(f"{word}: {score}")
```

### Task 4.3

```python
def classify_account(df, account_id, word_scores):
    account_score = 0
    account_tweets = df[df["account_id"] == account_id]
    
    if account_tweets.empty:
        return {"account_id": account_id, "account_type": "Unknown", "error": "Account not found"}
    
    for index, row in account_tweets.iterrows():
        words = row["text"]
        account_score += sum(word_scores.get(word, 0) for word in words)
        
    prediction = "Business actors" if account_score > 0 else "Private individuals"
    
    return {"account_id": account_id, "account_type": prediction, "account_score": account_score, "error": None}
```
