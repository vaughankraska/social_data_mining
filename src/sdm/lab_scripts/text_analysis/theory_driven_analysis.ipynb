```python
import pandas as pd
import krippendorff
```

```python
train_file_path = f'D:/UU/Sem3/SDM/social_data_mining/data/train.xlsx'
excel_data = pd.ExcelFile(train_file_path)
sheet_names = excel_data.sheet_names
print("Sheet names: ", sheet_names)
```

```python
train_coder1_df = pd.read_excel(train_file_path, sheet_name=sheet_names[1])
train_coder2_df = pd.read_excel(train_file_path, sheet_name=sheet_names[2])
train_coder3_df = pd.read_excel(train_file_path, sheet_name=sheet_names[3])
```

```python
train_merge_1_2_df = train_coder1_df.merge(train_coder2_df, on='ID', suffixes=("_1", "_2"))
train_merge_1_2_df = train_merge_1_2_df.drop(columns=["TEXT_2"]).rename(columns={"TEXT_1": "TEXT"})
train_merge_1_2_df.head()
```

```python
train_coder3_df_copy = train_coder3_df.copy().drop(columns=["TEXT"]).rename(columns={"CODE": "CODE_3"})
train_merge_1_2_3_df = train_merge_1_2_df.merge(train_coder3_df_copy, on='ID')
train_merge_1_2_3_df.head()
```
