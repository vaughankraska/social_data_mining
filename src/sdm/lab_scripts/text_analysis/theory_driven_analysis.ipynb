### Task 4.2

```python
from text_mining.utils import preprocess_tweets
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df['text'] = preprocess_tweets(df['tweet_text'])
df.head(10)
```

```python
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['text'])
word_counts = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
word_counts["account_type"] = df["account_type"]

word_counts.head()
```

```python
grouped_counts = word_counts.groupby("account_type").sum()
total_counts = grouped_counts.sum(axis=0)
print(total_counts)
```

```python
# Log-odds is used to score and rank words

alpha = 1   # Smoothning parameter
log_odds = np.log((grouped_counts + alpha) / (grouped_counts.sum(axis=1).values[:, None] - grouped_counts + alpha)) - \
           np.log((total_counts + alpha) / (total_counts.sum() - total_counts + alpha))
           
log_odds = pd.DataFrame(log_odds.T, index=vectorizer.get_feature_names_out(), columns=grouped_counts.index)
```

```python
log_odds.head()
```

```python
top_words = log_odds.apply(lambda x: x.nlargest(10).index, axis=0)
print(top_words)
```
